## 1. Load Environment Variables

```python
from dotenv import load_dotenv
import os

# Load environment variables
load_dotenv()

# Verify Azure OpenAI configuration
print("Azure OpenAI Configuration:")
print(f"Endpoint: {os.getenv('AZURE_OPENAI_ENDPOINT')}")
print(f"Deployment: {os.getenv('AZURE_OPENAI_DEPLOYMENT_NAME')}")
print(f"API Version: {os.getenv('AZURE_OPENAI_API_VERSION')}")
print(f"API Key configured: {'Yes' if os.getenv('AZURE_OPENAI_API_KEY') else 'No'}")
```

## 2. Test Basic Azure OpenAI Connection

```python
from langchain_openai import AzureChatOpenAI

# Create Azure OpenAI model instance
model = AzureChatOpenAI(
    azure_deployment=os.getenv("AZURE_OPENAI_DEPLOYMENT_NAME"),
    api_version=os.getenv("AZURE_OPENAI_API_VERSION"),
    temperature=0,
)

# Test the model
response = model.invoke("Say 'Hello from Azure OpenAI!' in a creative way.")
print(response.content)
```

## 3. Using the Helper Module

```python
from azure_openai_setup import get_azure_chat_model

# Get model using helper function
model = get_azure_chat_model(temperature=0.7)

# Test it
response = model.invoke("Explain LangChain in one sentence.")
print(response.content)
```
