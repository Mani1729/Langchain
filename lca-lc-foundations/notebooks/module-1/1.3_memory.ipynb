```python
import os
import sys
from dotenv import load_dotenv

# Add parent directory to path to import azure_openai_config
sys.path.append(os.path.join(os.path.dirname(os.path.abspath('__file__')), '..', '..'))

load_dotenv()

from azure_openai_config import setup_azure_openai, get_azure_openai_model
setup_azure_openai()
```

## No memory

```python
from langchain.agents import create_agent


agent = create_agent(
    get_azure_openai_model()
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Seán and my favourite colour is green")

response = agent.invoke(
    {"messages": [question]} 
)
```

```python
from pprint import pprint

pprint(response)
```

```python
question = HumanMessage(content="What's my favourite colour?")

response = agent.invoke(
    {"messages": [question]} 
)

pprint(response)
```

## Memory

```python
from langgraph.checkpoint.memory import InMemorySaver  


agent = create_agent(
    get_azure_openai_model(),
    checkpointer=InMemorySaver(),  
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Seán and my favourite colour is green")
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages": [question]},
    config,  
)
```

```python
pprint(response)
```

```python
question = HumanMessage(content="What's my favourite colour?")

response = agent.invoke(
    {"messages": [question]},
    config,  
)

pprint(response)
```
